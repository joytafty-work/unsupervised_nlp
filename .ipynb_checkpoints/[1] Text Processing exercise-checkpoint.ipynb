# Text Processing Exercise: Turning text into bag or words

### Exercise 1.1 Read text data (raw TED transcripts) from '5.ted' from './ted_mini/art_positive/' directory

```python
f = open('./ted_mini/art_positive/5.ted', 'r')
lines = f.readlines()
# print lines

import re
new_lines = re.sub('_en', '', lines[0])
# print lines[0].replace('_en', '')

### Create giant list
# a = []
# for x in lines: 
#     a.extent()
from itertools import chain

giant_list = list(chain(lines))
# print giant_list

clean_giant_list = [re.sub('_en', '', x) for x in giant_list]
clean_giant_list = map(lambda x: re.sub('_en', '', x), giant_list)

print type(clean_giant_list), len(clean_giant_list)

clean_giant_list[0]
```

### Exercise 1.2: Tokenize text into words

```python
tokens = list(chain(*map(lambda x: x.split(' '), clean_giant_list)))
print len(set(tokens))
print tokens
```

### Exercise 1.3: Remove Punctuations

```python
# Example: 
import re, string
pattern = '[{}]'.format(re.escape(string.punctuation))
my_str = ",Welcome! to the world of 'alphanumeric!!!- We remove punctuations!'"
print my_str
print re.sub('[{}]'.format(re.escape(string.punctuation)), '', my_str.lower())
print ' '.join(re.findall(r'\w+', my_str.lower()))
```

```python
## Hint 1: You can use string.punctuation to list all ASCII punctuation characters. (https://docs.python.org/2/library/string.html) 

## Hint 2: You can use regex pattern r'\w+' to match any alphanumeric and underscore (equivalent to the set [a-zA-Z0-9_]).
tokens = map(lambda x: re.sub('[{}]'.format(re.escape(string.punctuation)), '', x), tokens)
print tokens
```

### Exercise 1.4: Any Additional cleanup?

```python
## There are empty tokens and tokens that only contain digits. Let's remove them
tokens = filter(lambda x: (len(x) > 1) and (not x.isdigit()), tokens)
print tokens
```

### Exercise 1.5 Lemmatization

```python
# Hint: You can use the following libraries to lemmatize text
# 1. WordNet Lemmatizer: 
from nltk.stem.wordnet import WordNetLemmatizer
lm = WordNetLemmatizer()
token_lemmatized = map(lambda x: lm(x), tokens)

# 2. Porter Stemmer: 
from nltk.stem import PorterStemmer
st = PorterStemmer()
token_stemmed = map(lambda x: st(x), tokens)
```
